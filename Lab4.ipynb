```python
import pandas as pd
```

```python
data = pd.read_csv('Euro_2012_stats_TEAM.csv')
```

```python
data
```

## Petro

### Teams with shooting accuracy over average

```python
data['Shooting Accuracy num'] = data['Shooting Accuracy'].apply(lambda x: float(x[:-1]))
data[data['Shooting Accuracy num']  > data['Shooting Accuracy num'].mean()].Team
```

### Teams with more goals conceded than scored

```python
data[data['Goals conceded'] > data['Goals']].Team
```

### Team with max amount of failed passes

```python
data['Failed Passes'] = data['Passes'] - data['Passes completed']
data[data['Failed Passes'] == data['Failed Passes'].max()].Team
```

## Stas

### Which countries scored more than 8 goals? 

```python
data[data['Goals'] > 8].Team
```

### The country with the most fouls?

```python
data[data['Fouls Conceded'] == data['Fouls Conceded'].max()].Team
```

### How many fouls were commited by the Netherlands?

```python
data[data['Team'] == 'Netherlands']['Fouls Conceded']
```

## Andrii

### Countries with more fouls conceded than won

```python
data[data['Fouls Conceded'] < data['Fouls Won']].Team
```

### Countries with more than 18 players used

```python
data[data['Players Used'] > 18].Team
```

### Teams which had red cards and scored penaties

```python
data[(data['Red Cards'] > 0) & (data['Penalty goals'] > 0)].Team
```
