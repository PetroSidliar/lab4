```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Практика/Udemy_Finance.csv')
```

```python
data.head()
```

```python
#Перевірка 5 випадкових рядків, де курс безкоштовний 
df_price_is_paid = data[['price_detail__amount', 'is_paid']]
df_price_is_paid[df_price_is_paid['is_paid'] == False].sample(5)
```

```python
#Перевірка кількості рядків, де ціна зі знижкою не дорівнює 455, не NaN 
series_bool_discount_price = ((data['discount_price__amount'] != 455) & (data['discount_price__amount'].notna()))
not_455_not_na_count = series_bool_discount_price.sum()
not_455_not_na_count
```

```python
#Перевірка скільки курсів є платних, але їх ціна 0 
((data['is_paid']) & (data['price_detail__amount'].isna())).sum()
```

```python
#Знаходження курсу з недійсним введенням ціни
data[data['is_paid'] & data['price_detail__amount'].isna()]
```

```python
courses = pd.read_csv('Практика/Udemy_General.csv')
```

```python
courses
```

```python
#На які курси є найбільше зарахованих?
courses.sort_values(by = 'Enrollment', ascending=False).head(10)
```

```python
fig, ax = plt.subplots(figsize=(16,10))
courses.Category.value_counts().plot.pie(ax=ax, autopct='%1.2f%%', shadow=True, explode=(0.1, 0, 0, 0, 0), textprops={'size':'x-large', 'weight':'bold'})
plt.ylabel('')
plt.title('Скільки курсів у кожній категорії у відсотковому співвідношенні?', fontsize=20)
plt.show()
```
